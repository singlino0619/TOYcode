```python
import requests
from bs4 import BeautifulSoup
```

#### タイトルをスクレイピング
- requests.get()関数を用いてhtmlを取得
- beautiful形式で扱えるオブジェクト生成
    - *BeautifulSoup(*.text,'lxml')
- beutifulsoupオブジェクト. title . stringでhtmlの中のtitleタグを探し，stringで表示する
![エビフライトライアングル](scraping_extraxt_title.png)

```python
response = requests.get('https://suumo.jp/')
soup = BeautifulSoup(response.text, 'lxml')
title = soup.title.string
print(title)
```

#### リンク先をスクレイピング
- soup.a.get('href')でhtmlの中の最初のリンク先を取得
- すべてのリンク先を取得したい場合は
    - soup.findAll('a')のようにaタグにマッチするものを取ってくる
    - findAll('a')関数で取得したものは，配列になっているので，1行ずつ取り出したかったらfor文を用いる

```python
response = requests.get('https://suumo.jp/')
soup = BeautifulSoup(response.text, 'lxml')
links = soup.findAll('a')
for link in links:
    print(link.get('href'))
```

```python
response = requests.get('https://suumo.jp/tokyo/ensen/')
soup = BeautifulSoup(response.text, 'lxml')
links = soup.findAll('a')
href_links = ['' for i in range(len(links))]
for i in range(len(href_links)):
    href_links[i] = links[i].get('href')
```
